```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merge_df = pd.merge(city_df, ride_df, on="city")

# Display the data table for preview
merged_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_df = merge_df.loc[merge_df['type'] == "Urban"]
suburban_df = merge_df.loc[merge_df['type'] == "Suburban"]
rural_df = merge_df.loc[merge_df['type'] == "Rural"]
urban_group = urban_df.groupby(['city'])
suburban_group = suburban_df.groupby(['city'])
rural_group = rural_df.groupby(['city'])

urban_driver_count = urban_group.ride_id.count()
suburban_driver_count = suburban_group.ride_id.count()
rural_driver_count = rural_group.ride_id.count()
urban_total_fares = urban_group.fare.sum()
suburban_total_fares = suburban_group.fare.sum()
rural_total_fares = rural_group.fare.sum()

urban_avg_fare = urban_total_fares / urban_driver_count
suburban_avg_fare = suburban_total_fares / suburban_driver_count
rural_avg_fare = rural_total_fares / rural_driver_count

# Build the scatter plots for each city types
city_df = city_df.set_index(['city'])
city_new_df = city_df.sort_index()

urban_df = city_new_df.loc[city_new_df['type'] == "Urban"]
suburban_df = city_new_df.loc[city_new_df['type'] == "Suburban"]
rural_df = city_new_df.loc[city_new_df['type'] == "Rural"]

urban_bubble = urban_df['driver_count'] * 10
suburban_bubble = suburban_df['driver_count'] * 10
rural_bubble = rural_df['driver_count'] * 10

urban_handle = plt.scatter(urban_driver_count, urban_avg_fare, marker="o", c='orange', edgecolors="black", s=urban_bubble, alpha=0.75, label="Urban")
suburban_handle = plt.scatter(suburban_driver_count, suburban_avg_fare, marker="o", c='lightskyblue', edgecolors="black", s=suburban_bubble, alpha=0.75, label="Suburban")
rural_handle = plt.scatter(rural_driver_count, rural_avg_fare, marker="o", c='gold', edgecolors="black", s=rural_bubble, alpha=0.75, label="Rural")

# Incorporate the other graph properties
plt.xlim(0, 40)
plt.ylim(10, 45)
plt.grid(axis='both', alpha=0.5)
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

# Create a legend
lgnd = plt.legend(loc="upper right", scatterpoints=1, fontsize=10)
lgnd.legendHandles[0]._sizes = [40]
lgnd.legendHandles[1]._sizes = [40]
lgnd.legendHandles[2]._sizes = [40]
lgnd.set_title("City Types", prop = {'size':'large'})

# Incorporate a text label regarding circle size
plt.text(x=42, y=35, s='Note:\nCircle size correlates to driver count per city.')

# Save Figure
plt.savefig("pyber_starter_plot.png", bbox_inches="tight")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fares = merge_df['fare'].sum()
type_group = merge_df.groupby('type')
fares_per_type = type_group.fare.sum()
type_percent = (fares_per_type * 100) / total_fares

# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
sizes = [6.8, 30.5, 62.7]
colors = ["gold", "lightskyblue", "orange"]
explode = (0, 0, 0.08)

# Save Figure
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%2.2f%%", shadow=True, startangle=160)

plt.savefig("Total_Fares_by_City_Type.png")

plt.title('% of Total Fares by City Type')

```
